```python
import pandas as pd

arquivo = pd.read_excel("cursinho.xlsx")
display(arquivo)
```

```python
import matplotlib.pyplot as plt

mat = arquivo["Matérias"]
alu = arquivo["Alunos presentes"] 



plt.title("Porcentagem de frequência dos alunos nas matérias")
plt.pie(alu, labels=mat, autopct="%1.2f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt

mat = arquivo["Matérias"]
alu = arquivo["Alunos com falta"]



plt.title("Porcentagem de Ausência dos alunos nas matérias")
plt.pie(alu, labels=mat, autopct="%1.2f%%")
plt.show()
```

```python

# Agrupar por dia da semana e somar as presenças e faltas
total_presencas = arquivo.groupby('Dias da semana')['Alunos presentes'].sum()
total_faltas = arquivo.groupby('Dias da semana')['Alunos com falta'].sum()

# Plotar as presenças e faltas
plt.figure(figsize=(10, 5))

plt.plot(total_presencas.index, total_presencas.values, marker='o', label='Presenças', color='green')
plt.plot(total_faltas.index, total_faltas.values, marker='o', label='Faltas', color='red')

# Configurações do gráfico
plt.title('Presenças e Faltas por Dias da Semana')
plt.xlabel('Dias da Semana')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
total_por_materia = arquivo.groupby('Matérias')[['Alunos presentes', 'Alunos com falta']].sum()

# Plotar o gráfico de barras
total_por_materia.plot(kind='bar', figsize=(10, 5), color=['green', 'red'])

# Configurações do gráfico
plt.title('Frequência por Matéria: Presenças e Faltas')
plt.xlabel('Matérias')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.legend(title='Tipo')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```

```python
# Agrupar por dia da semana e somar as presenças e faltas
total_presencas = arquivo.groupby('Dias da semana')['Alunos presentes'].sum()
total_faltas = arquivo.groupby('Dias da semana')['Alunos com falta'].sum()

# Encontrar o dia com mais presenças
dia_mais_presencas = total_presencas.idxmax()
presencas_maximas = total_presencas.max()

# Encontrar o dia com mais faltas
dia_mais_faltas = total_faltas.idxmax()
faltas_maximas = total_faltas.max()

print(f"O dia com mais presenças é {dia_mais_presencas} com {presencas_maximas} presenças.")
print(f"O dia com mais faltas é {dia_mais_faltas} com {faltas_maximas} faltas.")
```
